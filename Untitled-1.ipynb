````python
import numpy as np
import pandas as pd

# Khởi tạo dữ liệu
data = np.array([13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70])

# Tính trung bình
mean = np.mean(data)
print("Trung bình: ", mean)

# Tính trung vị
median = np.median(data)
print("Trung vị: ", median)

# Tính mode
mode = pd.Series(data).mode()
print("Mode: ", mode.values)

# Tính trung điểm
midrange = (np.max(data) + np.min(data)) / 2
print("Trung điểm: ", midrange)

# Tính tứ phân vị
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
print("Tứ phân vị thứ nhất (Q1): ", q1)
print("Tứ phân vị thứ ba (Q3): ", q3)

# Tóm tắt năm số liệu
summary = pd.DataFrame(data).describe(percentiles=[.25, .5, .75])
print("Tóm tắt năm số liệu: \n", summary)

# Vẽ biểu đồ hộp
import matplotlib.pyplot as plt
plt.boxplot(data)
plt.show()
````

Để tính trung vị xấp xỉ cho dữ liệu được nhóm thành các khoảng, ta có thể sử dụng phương pháp ước lượng trung vị bằng cách sử dụng công thức:

Median = L + [(n/2 - F)/f] x w

Trong đó:

L: giá trị trung bình của khoảng chứa trung vị
n: tổng số quan sát
F: tần số tích lũy của khoảng trước khoảng chứa trung vị
f: tần số của khoảng chứa trung vị
w: độ rộng của khoảng chứa trung vị
Áp dụng công thức trên, ta có:

L = 35.5 (khoảng 21-50 là khoảng chứa trung vị và trung bình của khoảng này là (21+50)/2 = 35.5)
n = 3194 (tổng số quan sát)
F = 950 (tần số tích lũy của khoảng 16-20)
f = 1500 (tần số của khoảng 21-50)
w = 30 (độ rộng của khoảng 21-50)
Thay các giá trị vào công thức, ta có:

Median = 35.5 + [(3194/2 - 950)/1500] x 30
= 35.5 + [622/1500] x 30
= 35.5 + 12.44
≈ 47.94

Vậy trung vị xấp xỉ của dữ liệu là khoảng 48.

````python
# Dữ liệu về tuổi và %mỡ
age = np.array([23, 23, 27, 27, 39, 41, 47, 49, 50, 52, 54, 54, 56, 57, 58, 58, 60, 61])
body_fat = np.array([9.5, 26.5, 7.8, 17.8, 31.4, 25.9, 27.4, 27.2, 31.2, 34.6, 42.5, 28.8, 33.4, 30.2, 34.1, 32.9, 41.2, 35.7])

# Tính trung bình
age_mean = np.mean(age)
body_fat_mean = np.mean(body_fat)

# Tính trung vị
age_median = np.median(age)
body_fat_median = np.median(body_fat)

# Tính độ lệch chuẩn
age_std = np.std(age)
body_fat_std = np.std(body_fat)

print("Tuổi:")
print("Trung bình:", age_mean)
print("Trung vị:", age_median)
print("Độ lệch chuẩn:", age_std)

print("%mỡ:")
print("Trung bình:", body_fat_mean)
print("Trung vị:", body_fat_median)
print("Độ lệch chuẩn:", body_fat_std)
````

- Khoảng cách Manhattan: $\sum_{i=1}^{n}|x_i - y_i|$

a) Để xếp hạng các điểm dữ liệu trong cơ sở dữ liệu dựa trên độ tương đồng với truy vấn theo khoảng cách Euclid, khoảng cách Manhattan, khoảng cách supremum và độ tương đồng cos, ta tính khoảng cách giữa điểm truy vấn và mỗi điểm dữ liệu trong cơ sở dữ liệu bằng các thước đo tương ứng và sắp xếp các điểm dữ liệu theo thứ tự tăng dần của khoảng cách.

- Khoảng cách Euclid giữa điểm truy vấn và mỗi điểm dữ liệu:

    $d_{Euclid}(x, x_i) = \sqrt{(x_1 - x_{i1})^2 + (x_2 - x_{i2})^2}$

    Với điểm truy vấn $x = (1.4, 1.6)$ và các điểm dữ liệu trong cơ sở dữ liệu, ta tính được:

    $d_{Euclid}(x, x_1) = \sqrt{(1.4 - 1.5)^2 + (1.6 - 1.7)^2} \approx 0.141$
    
    $d_{Euclid}(x, x_2) = \sqrt{(1.4 - 2)^2 + (1.6 - 1.9)^2} \approx 0.447$
    
    $d_{Euclid}(x, x_3) = \sqrt{(1.4 - 1.6)^2 + (1.6 - 1.8)^2} \approx 0.223$
    
    $d_{Euclid}(x, x_4) = \sqrt{(1.4 - 1.2)^2 + (1.6 - 1.5)^2} \approx 0.282$
    
    $d_{Euclid}(x, x_5) = \sqrt{(1.4 - 1.5)^2 + (1.6 - 1)^2} \approx 0.640$

    Vậy thứ tự các điểm dữ liệu từ gần đến xa với điểm truy vấn theo khoảng cách Euclid là: $x_1$, $x_3$, $x_4$, $x_2$, $x_5$.

- Khoảng cách Manhattan giữa điểm truy vấn và mỗi điểm dữ liệu:

    $d_{Manhattan}(x, x_i) = |x_1 - x_{i1}| + |x_2 - x_{i2}|$

    Với điểm truy vấn $x = (1.4, 1.6)$ và các điểm dữ liệu trong cơ sở dữ liệu, ta tính được:

    $d_{Manhattan}(x, x_1) = |1.4 - 1.5| + |1.6 - 1.7| = 0.2$
    
    $d_{Manhattan}(x, x_2) = |1.4 - 2| + |1.6 - 1.9| = 0.9$
    
    $d_{Manhattan}(x, x_3) = |1.4 - 1.6| + |1.6 - 1.8| = 0.4$
    
    $d_{Manhattan}(x, x_4) = |1.4 - 1.2| + |1.6 - 1.5| = 0.3$
    
    $d_{Manhattan}(x, x_5) = |1.4 - 1.5| + |1.6 - 1| = 0.5$

    Vậy thứ tự các điểm dữ liệu từ gần đến xa với điểm truy vấn theo khoảng cách Manhattan là: $x_1$, $x_4$, $x_3$, $x_5$, $x_2$.

- Khoảng cách supremum giữa điểm truy vấn và mỗi điểm dữ liệu:

    $d_{supremum}(x, x_i) = \max(|x_1 - x_{i1}|, |x_2 - x_{i2}|)$

    Với điểm truy vấn $x = (1.4, 1.6)$ và các điểm dữ liệu trong cơ sở dữ liệu, ta tính được:

    $d_{supremum}(x, x_1) = \max(|1.4 - 1.5|, |1.6 - 1.7|) = 0.1$
    
    $d_{supremum}(x, x_2) = \max(|1.4 - 2|, |1.6 - 1.9|) = 0.6$
    
    $d_{supremum}(x, x_3) = \max(|1.4 - 1.6|, |1.6 - 1.8|) = 0.2$
    
    $d_{supremum}(x, x_4) = \max(|1.4 - 1.2|, |1.6 - 1.5|) = 0.4$
    
    $d_{supremum}(x, x_5) = \max(|1.4 - 1.5|, |1.6 - 1|) = 0.6$

    Vậy thứ tự các điểm dữ liệu từ gần đến xa với điểm truy vấn theo khoảng cách supremum là: $x_1$, $x_3$, $x_4$, $x_2$, $x_5$.

- Độ tương đồng cos giữa điểm truy vấn và mỗi điểm dữ liệu:

    $sim_{cos}(x, x_i) = \frac{\sum_{j=1}^d x_j x_{ij}}{\sqrt{\sum_{j=1}^d x_j^2} \sqrt{\sum_{j=1}^d x_{ij}^2}}$

    Với điểm truy vấn $x = (1.4, 1.6)$ và các điểm dữ liệu trong cơ sở dữ liệu, ta tính được:

    $sim_{cos}(x, x_2) = \frac{1.4 \times 2 + 1.6 \times 1.9}{\sqrt{1.4^2 + 1.6^2} \sqrt{2^2 + 1.9^2}} \approx 0.986$
    
    $sim_{cos}(x, x_3) = \frac{1.4 \times 1.6 + 1.6 \times 1.8}{\sqrt{1.4^2 + 1.6^2} \sqrt{1.6^2 + 1.8^2}} \approx 0.997$

    $sim_{cos}(x, x_5) = \frac{1.4 \times 1.5 + 1.6 \times 1}{\sqrt{1.4^2 + 1.6^2} \sqrt{1.5^2 + 1^2}} \approx 0.911$

    Vậy thứ tự các điểm dữ liệu từ giống nhất đến khác nhất với điểm truy vấn theo độ tương đồng cos là: $x_1$, $x_3$, $x_2$, $x_4$, $x_5$.

a) Chuẩn hóa min-max bằng cách điều chỉnh $new\_min = 0$ và $new\_max = 1$:

Để chuẩn hóa min-max, ta sử dụng công thức sau:

$$x_{new} = \frac{x - min(x)}{max(x) - min(x)} \times (new\_max - new\_min) + new\_min$$

Trong đó:
- $x$ là giá trị ban đầu của dữ liệu
- $min(x)$ là giá trị nhỏ nhất trong dữ liệu
- $max(x)$ là giá trị lớn nhất trong dữ liệu
- $new\_min$ và $new\_max$ là giá trị tối thiểu và tối đa mới mà ta muốn chuẩn hóa dữ liệu về.

Áp dụng công thức trên, ta có:

$$200_{new} = \frac{200 - 200}{1000 - 200} \times (1 - 0) + 0 = 0$$

$$300_{new} = \frac{300 - 200}{1000 - 200} \times (1 - 0) + 0 = 0.111$$

$$400_{new} = \frac{400 - 200}{1000 - 200} \times (1 - 0) + 0 = 0.222$$

$$600_{new} = \frac{600 - 200}{1000 - 200} \times (1 - 0) + 0 = 0.444$$

$$1000_{new} = \frac{1000 - 200}{1000 - 200} \times (1 - 0) + 0 = 1$$

Vậy dữ liệu sau khi chuẩn hóa min-max là: 0, 0.111, 0.222, 0.444, 1.

b) Chuẩn hóa z:

Để chuẩn hóa z, ta sử dụng công thức sau:

$$x_{new} = \frac{x - \bar{x}}{s}$$

Trong đó:
- $x$ là giá trị ban đầu của dữ liệu
- $\bar{x}$ là giá trị trung bình của dữ liệu
- $s$ là độ lệch chuẩn của dữ liệu.

Đầu tiên, ta tính giá trị trung bình và độ lệch chuẩn của dữ liệu:

$$\bar{x} = \frac{200 + 300 + 400 + 600 + 1000}{5} = 500$$

$$s = \sqrt{\frac{(200 - 500)^2 + (300 - 500)^2 + (400 - 500)^2 + (600 - 500)^2 + (1000 - 500)^2}{5}} \approx 264.58$$

Áp dụng công thức chuẩn hóa z, ta có:

$$200_{new} = \frac{200 - 500}{264.58} \approx -1.13$$

$$300_{new} = \frac{300 - 500}{264.58} \approx -0.75$$

$$400_{new} = \frac{400 - 500}{264.58} \approx -0.38$$

$$600_{new} = \frac{600 - 500}{264.58} \approx 0.38$$

$$1000_{new} = \frac{1000 - 500}{264.58} \approx 1.51$$

Vậy dữ liệu sau khi chuẩn hóa z là: -1.13, -0.75, -0.38, 0.38, 1.51.

c) Chuẩn hóa z dùng trung bình độ lệch tuyệt đối thay vì độ lệch chuẩn:

Để chuẩn hóa z dùng trung bình độ lệch tuyệt đối, ta sử dụng công thức sau:

$$x_{new} = \frac{x - \bar{x}}{MAD}$$

Trong đó:
- $x$ là giá trị ban đầu của dữ liệu
- $\bar{x}$ là giá trị trung bình của dữ liệu
- $MAD$ là độ lệch tuyệt đối trung bình của dữ liệu.

Đầu tiên, ta tính giá trị trung bình và độ lệch tuyệt đối trung bình của dữ liệu:

$$\bar{x} = \frac{200 + 300 + 400 + 600 + 1000}{5} = 500$$

$$MAD = \frac{|200 - 500| + |300 - 500| + |400 - 500| + |600 - 500| + |1000 - 500|}{5} = 200$$

Áp dụng công thức chuẩn hóa z dùng trung bình độ lệch tuyệt đối, ta có:

$$200_{new} = \frac{200 - 500}{200} = -1.5$$

$$300_{new} = \frac{300 - 500}{200} = -1$$

$$400_{new} = \frac{400 - 500}{200} = -0.5$$

$$600_{new} = \frac{600 - 500}{200} = 0.5$$

$$1000_{new} = \frac{1000 - 500}{200} = 2.5$$

Vậy dữ liệu sau khi chuẩn hóa z dùng trung bình độ lệch tuyệt đối là: -1.5, -1, -0.5, 0.5, 2.5.

d) Chuẩn hóa bằng thang thập phân:

Để chuẩn hóa bằng thang thập phân, ta sử dụng công thức sau:

$$x_{new} = \frac{x - \lfloor x \rfloor}{1}$$

Trong đó:
- $x$ là giá trị ban đầu của dữ liệu
- $\lfloor x \rfloor$ là phần nguyên của $x$.

Áp dụng công thức chuẩn hóa bằng thang thập phân, ta có:

$$200_{new} = \frac{200 - \lfloor 200 \rfloor}{1} = 0$$

$$300_{new} = \frac{300 - \lfloor 300 \rfloor}{1} = 0.1$$

$$400_{new} = \frac{400 - \lfloor 400 \rfloor}{1} = 0.2$$

$$600_{new} = \frac{600 - \lfloor 600 \rfloor}{1} = 0.4$$

$$1000_{new} = \frac{1000 - \lfloor 1000 \rfloor}{1} = 0$$

Vậy dữ liệu sau khi chuẩn hóa bằng thang thập phân là: 0, 0.1, 0.2, 0.4, 0.

- Sử dụng phương pháp chuẩn hóa z để biến đổi giá trị 35 của thuộc tính tuổi, trong đó độ lệch chuẩn của tuổi là 12.70 năm:
    $$x_{new} = \frac{x - \bar{x}}{s}$$

    $$s = 12.70$$
    $$35_{new} = \frac{35 - 29.96}{12.70} \approx 0.40$$
    Giá trị 35 sau khi chuẩn hóa sẽ có giá trị z-score là 0.40.

- Sử dụng phương pháp chuẩn hóa bằng thang thập phân để biến đổi giá trị 35 cho thuộc tính tuổi:

b) Tính hệ số tương quan (hệ số Pearson) và hiệp phương sai của hai thuộc tính:
Để tính hệ số tương quan giữa hai thuộc tính tuổi và %mỡ cơ thể, ta sử dụng công thức:
$$r = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n}(x_i - \bar{x})^2}\sqrt{\sum_{i=1}^{n}(y_i - \bar{y})^2}}$$
Trong đó, $\bar{x}$ và $\bar{y}$ là giá trị trung bình của thuộc tính tuổi và %mỡ cơ thể, $x_i$ và $y_i$ là giá trị của thuộc tính tuổi và %mỡ cơ thể của quan sát thứ $i$.

Tính giá trị trung bình và độ lệch chuẩn của hai thuộc tính như trên, ta có:

$$r = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n}(x_i - \bar{x})^2}\sqrt{\sum_{i=1}^{n}(y_i - \bar{y})^2}} = \frac{0.82}{\sqrt{1.77}\sqrt{0.94}} \approx 0.89$$

Hệ số tương quan giữa hai thuộc tính tuổi và %mỡ cơ thể là 0.89, cho thấy một mối tương quan mạnh giữa hai thuộc tính này. Hệ số tương quan dương cho thấy khi tuổi tăng thì %mỡ cơ thể cũng tăng, và ngược lại.

$$cov(x,y) = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{n-1}$$

- Tần suất xuất hiện của $hamburgers$: 2000
- Tần suất xuất hiện của cả $hot\; dogs$ và $hamburgers$: 2000

- Support của $hot\; dogs$: 3000/5000 = 0.6
- Support của $hamburgers$: 2000/5000 = 0.4
- Support của cả $hot\; dogs$ và $hamburgers$: 2000/5000 = 0.4
- Confidence của luật kết hợp $hot\; dogs \Rightarrow hamburgers$: 0.4/0.6 = 0.67
- Lift của luật kết hợp $hot\; dogs \Rightarrow hamburgers$: (0.4/0.6) / 0.4 = 1.11

- all_confidence: Tính toán trung bình của tất cả các confidence của các quy tắc kết hợp chứa một mục cụ thể. Không áp dụng được trong trường hợp chỉ có hai mục.
- max_confidence: Chỉ lấy confidence cao nhất của các quy tắc kết hợp chứa một mục cụ thể. Không áp dụng được trong trường hợp chỉ có hai mục.
- Kulczynski: Tính toán trung bình của hai giá trị: $\frac{1}{2}(\frac{supp(X \Rightarrow Y)}{supp(X)} + \frac{supp(X \Rightarrow Y)}{supp(Y)})$. Giá trị này càng cao thì mức độ mạnh của luật càng cao.
- cosine: Tính toán cosine của vector tần suất của các mục trong các giao dịch. Giá trị này càng gần 1 thì mức độ mạnh của luật càng cao.
- lift: Tính toán tỷ lệ giữa support của cả hai mục và tích của support của từng mục riêng lẻ. Giá trị này càng lớn hơn 1 thì mức độ tương quan giữa hai mục càng cao.

````python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import graphviz

# Đọc dữ liệu từ file csv
data = pd.read_csv('bai6.1.csv')

# Tách dữ liệu thành features và target
X = data.drop('buys_computer', axis=1)
y = data['buys_computer']

# Áp dụng mã hóa one-hot cho các thuộc tính chuỗi
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)

# Lấy danh sách tên thuộc tính hoàn chỉnh
cat_feature_names = encoder.categories_
cat_feature_names = [f"{col}_{val}" for col, vals in zip(X.select_dtypes(include='object').columns, cat_feature_names) for val in vals]
num_feature_names = X.select_dtypes(include='number').columns
feature_names = list(num_feature_names) + list(cat_feature_names)

# Khởi tạo mô hình cây quyết định
clf = DecisionTreeClassifier()

# Huấn luyện mô hình trên dữ liệu đã được mã hóa
clf.fit(X_encoded, y)

# Dự đoán nhãn cho một mẫu mới
new_data = pd.DataFrame({'age': ['youth'], 'income': ['medium'], 'student': ['no'], 'credit_rating': ['fair']})
new_data_encoded = encoder.transform(new_data)

prediction = clf.predict(new_data_encoded)

print(prediction)

# Vẽ đồ thị cây quyết định
dot_data = export_graphviz(clf, out_file=None, feature_names=feature_names, class_names=['no', 'yes'], filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render('decision_tree', format='png', view=True)
````
