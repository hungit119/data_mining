- Khong cch Minkowski: $(\sum_{i=1}^{n}|x_i - y_i|^h)^{\frac{1}{h}}$
- Khong cch supremum: $\max_{i=1}^{n}|x_i - y_i|$

a) Chun ha min-max bng cch iu chnh $new\_min = 0$ v $new\_max = 1$:

 chun ha min-max, ta s dng cng thc sau:

$$x_{new} = \frac{x - min(x)}{max(x) - min(x)} \times (new\_max - new\_min) + new\_min$$

Trong :
- $x$ l gi tr ban u ca d liu
- $min(x)$ l gi tr nh nht trong d liu
- $max(x)$ l gi tr ln nht trong d liu
- $new\_min$ v $new\_max$ l gi tr ti thiu v ti a mi m ta mun chun ha d liu v.

p dng cng thc trn, ta c:

$$200_{new} = \frac{200 - 200}{1000 - 200} \times (1 - 0) + 0 = 0$$

$$300_{new} = \frac{300 - 200}{1000 - 200} \times (1 - 0) + 0 = 0.111$$

$$400_{new} = \frac{400 - 200}{1000 - 200} \times (1 - 0) + 0 = 0.222$$

$$600_{new} = \frac{600 - 200}{1000 - 200} \times (1 - 0) + 0 = 0.444$$

$$1000_{new} = \frac{1000 - 200}{1000 - 200} \times (1 - 0) + 0 = 1$$

Vy d liu sau khi chun ha min-max l: 0, 0.111, 0.222, 0.444, 1.

- $x$ l gi tr ban u ca d liu
- $\bar{x}$ l gi tr trung bnh ca d liu
- $s$ l  lch chun ca d liu.

$$400_{new} = \frac{400 - 500}{264.58} \approx -0.38$$

$$600_{new} = \frac{600 - 500}{264.58} \approx 0.38$$

$$1000_{new} = \frac{1000 - 500}{264.58} \approx 1.51$$

Vy d liu sau khi chun ha z l: -1.13, -0.75, -0.38, 0.38, 1.51.

Trong :
- $x$ l gi tr ban u ca d liu
- $\bar{x}$ l gi tr trung bnh ca d liu
- $MAD$ l  lch tuyt i trung bnh ca d liu.

b) Tnh h s tng quan (h s Pearson) v hip phng sai ca hai thuc tnh:
 tnh h s tng quan gia hai thuc tnh tui v %m c th, ta s dng cng thc:
$$r = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n}(x_i - \bar{x})^2}\sqrt{\sum_{i=1}^{n}(y_i - \bar{y})^2}}$$
Trong , $\bar{x}$ v $\bar{y}$ l gi tr trung bnh ca thuc tnh tui v %m c th, $x_i$ v $y_i$ l gi tr ca thuc tnh tui v %m c th ca quan st th $i$.

Tnh gi tr trung bnh v  lch chun ca hai thuc tnh nh trn, ta c:

$$r = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n}(x_i - \bar{x})^2}\sqrt{\sum_{i=1}^{n}(y_i - \bar{y})^2}} = \frac{0.82}{\sqrt{1.77}\sqrt{0.94}} \approx 0.89$$

H s tng quan gia hai thuc tnh tui v %m c th l 0.89, cho thy mt mi tng quan mnh gia hai thuc tnh ny. H s tng quan dng cho thy khi tui tng th %m c th cng tng, v ngc li. 

 tnh hip phng sai ca hai thuc tnh tui v %m c th, ta s dng cng thc:
$$cov(x,y) = \frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}{n-1}$$
p dng cng thc trn, ta tnh c hip phng sai ca hai thuc tnh l 16.96.

Tm li, hai thuc tnh tui v %m c th c mi tng quan mnh vi h s tng quan dng l 0.89 v hip phng sai l 16.96.

````python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import graphviz

# c d liu t file csv
data = pd.read_csv('bai6.1.csv')

# Tch d liu thnh features v target
X = data.drop('buys_computer', axis=1)
y = data['buys_computer']

# p dng m ha one-hot cho cc thuc tnh chui
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)

# Ly danh sch tn thuc tnh hon chnh
cat_feature_names = encoder.categories_
cat_feature_names = [f"{col}_{val}" for col, vals in zip(X.select_dtypes(include='object').columns, cat_feature_names) for val in vals]
num_feature_names = X.select_dtypes(include='number').columns
feature_names = list(num_feature_names) + list(cat_feature_names)

# Khi to m hnh cy quyt nh
clf = DecisionTreeClassifier()

# Hun luyn m hnh trn d liu  c m ha
clf.fit(X_encoded, y)

# D on nhn cho mt mu mi
new_data = pd.DataFrame({'age': ['youth'], 'income': ['medium'], 'student': ['no'], 'credit_rating': ['fair']})
new_data_encoded = encoder.transform(new_data)

prediction = clf.predict(new_data_encoded)

print(prediction)

# V  th cy quyt nh
dot_data = export_graphviz(clf, out_file=None, feature_names=feature_names, class_names=['no', 'yes'], filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph.render('decision_tree', format='png', view=True)
````
